##### Creating train test split for data modelling and testing

```python
X = df.drop('benign_0__mal_1', axis=1).values
y = df['benign_0__mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

Taking a smaller test_size since the dataset is relatively small with only 569 entries

```python
# Scaling the dataset for more precise processing of the data
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

We are not fitting the scaler on the X_test to prevent data leakage. Training set should not in any way know about the data in the test_set

##### Creating ANN for model creation an d model testing

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
X_train.shape
```

```python
model = Sequential()
model.add(Dense(30, activation='relu'))
model.add(Dense(15, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

Compiling and fitting the Model

```python
model.compile(loss='binary_crossentropy', optimizer='adam')
model.fit(x=X_train, y=y_train, epochs=300, validation_data=(X_test, y_test))
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

We can observe that we seem to get close to the actual loss value but then digress from the path. In the next steps we will be applying EarlyStopping by monitoring 'val_loss' for being minimum.

```python
# Creating the model and adding Early Stopping Mechanism 
model = Sequential()
model.add(Dense(30, activation='relu'))
model.add(Dense(15, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop  = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=25)
```

patience is set to 25, the nuumber of epochs with no improvement after which training will be stopped.

Fitting the model with early stop callback

```python
model.fit(x=X_train, y=y_train, epochs=300, validation_data=(X_test, y_test), callbacks=[early_stop])
```

```python
model_loss = pd.DataFrame(model.history.history)
```

```python
model_loss.plot()
```

We were able to minimize the distance between the lines of 'val_loss' and 'loss' by using Early Stop callback but we can still see some scope of improvement. In the next steps we will add Dropout layers in our neural network. Dropout layer is the way to tell the NN to shut off the given number of neurons so that their weights and biases don't get updated

```python
from tensorflow.keras.layers import Dropout
```

```python
model = Sequential()
model.add(Dense(30, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(15, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.fit(x=X_train, y=y_train, epochs=600, validation_data=(X_test, y_test), callbacks=[early_stop])
```

```python
model_d_loss = pd.DataFrame(model.history.history)
```

```python
model_d_loss.plot()
```

We were able to minimize the distance between the lines of the loss and val_loss with very little scope of improvement. To verfiy that we will see the predictions, classification_report and confusion matrix

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
predictions = (model.predict(X_test) > 0.5).astype("int32")
```

```python
print(classification_report(y_test, predictions))
```

```python
print(confusion_matrix(y_test, predictions))
```

According to the classification report our model is predicting with the accuracy of 0.97 or 97%.
